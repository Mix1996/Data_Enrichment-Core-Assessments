```python
import pandas as pd
import ast
```

```python
superhero_info = pd.read_csv('/Users/MiPillay1/OneDrive - MMI HOLDINGS LTD/Documents/GitHub/Data_Enrichment-Core-Assessments/Data/superhero_info.csv')
superhero_powers = pd.read_csv('/Users/MiPillay1/OneDrive - MMI HOLDINGS LTD/Documents/GitHub/Data_Enrichment-Core-Assessments/Data/superhero_powers.csv')
```

```python
superhero_info.head()
```

```python
superhero_info = pd.DataFrame(superhero_info)
```

```python
# Split the 'Hero|Publisher' column into multiple columns
split_data = superhero_info['Hero|Publisher'].str.split('|', n=1, expand=True)
```

```python
# Assign the first split to 'Hero'
superhero_info['Hero'] = split_data[0]

# Check if there is a second split before assigning it to 'Publisher'
if len(split_data.columns) > 1:
    superhero_info['Publisher'] = split_data[1]
else:
    superhero_info['Publisher'] = None
```

```python
# Drop the original 'Hero|Publisher' column
superhero_info.drop('Hero|Publisher', axis=1, inplace=True)
```

```python
# Create a DataFrame from the sample data
superhero_powers = pd.DataFrame(superhero_powers)

# Split the 'Powers' column to create one-hot-encoded columns for each power
powers_encoded = superhero_powers['Powers'].str.get_dummies(',')

# Concatenate the one-hot-encoded powers with the 'hero_names' column
final_powers_df = pd.concat([superhero_powers['hero_names'], powers_encoded], axis=1)
```

```python
final_powers_df.head()
```

```python
# Merge the DataFrames on the 'Name' and 'Hero name' columns
final_df = superhero_info.merge(final_powers_df, left_on='Hero', right_on='hero_names', how='left')
```

```python
# Convert the 'Measurements' strings to dictionaries
final_df['Measurements'] = final_df['Measurements'].apply(ast.literal_eval)

# Extract numeric values from 'Height' and 'Weight' in the 'Measurements' column
final_df['Height'] = final_df['Measurements'].apply(lambda x: float(x['Height'].split(' ')[0]))
final_df['Weight'] = final_df['Measurements'].apply(lambda x: float(x['Weight'].split(' ')[0]))
```

```python
final_df.head()
```

```python
# Drop the original 'Measurements' column
final_df.drop('Measurements', axis=1, inplace=True)
```

```python
# Drop the original 'hero_names' column
final_df.drop('hero_names', axis=1, inplace=True)
```

# Compare the average weight of superheroes who have Super Speed to those who do not:

```python
# Calculate the average weight of superheroes with Super Speed
avg_weight_with_super_speed = final_df[final_df['Super Speed'] == 1]['Weight'].mean()

# Calculate the average weight of superheroes without Super Speed
avg_weight_without_super_speed = final_df[final_df['Super Speed'] == 0]['Weight'].mean()

# Print the results
print("Average weight of superheroes with Super Speed:", avg_weight_with_super_speed)
print("Average weight of superheroes without Super Speed:", avg_weight_without_super_speed)
```

# What is the average height of heroes for each publisher:

```python
# Calculate the average height of heroes for each publisher
average_height_by_publisher = final_df.groupby('Publisher')['Height'].mean()

# Print the results
print("Average height of heroes for each publisher:")
print(average_height_by_publisher)
```
